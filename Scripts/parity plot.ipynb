```python
import matplotlib.pyplot as plt
import matplotlib
import pandas as pd
import numpy as np
```

```python
# matplotlib.rcParams['font.family'] = "Arial"
df = pd.read_csv('../test_results.csv')
target = df.iloc[:,0].to_list()
pred = df.iloc[:,1].to_list()

fig,ax=plt.subplots(1,1,figsize=(10,10))
[ax.spines[n].set_linewidth(2) for n in ax.spines.keys()]

# x_data = scaler.inverse_transform()
ax.set_xticks(range(-5,4,1))
ax.set_yticks(range(-5,4,1))
ax.plot(np.arange(-20,20,1), np.arange(-20,20,1), c='k', lw = 1.4,ls='--', alpha= 0.8)
ax.scatter(target,pred,  s=70, c='limegreen', alpha=0.7)
ax.set_xlabel(r'Actual Value',fontsize=30, labelpad=10)
ax.set_ylabel(r'Predicted Value', fontsize=30, labelpad = 10)
ax.tick_params(labelsize=20, length = 7, width=3)
ax.set_xlim(-4,2)
ax.set_ylim(-4,2)
```
